```python
open("sobrepython.txt", "w")
# Abre o arquivo "sobrepython.txt" no modo de escrita ("w").
# Toda vez que um arquivo é aberto no modo de escrita, o conteúdo existente é sobrescrito. Se o arquivo não existir, ele será criado.
open("sobrepython.txt", "r")
# Abre o arquivo "sobrepython.txt" no modo de leitura ("r").
open("sobrepython.txt", "a")
# Abre o arquivo "sobrepython.txt" no modo de anexação ("a"), permitindo adicionar conteúdo ao final do arquivo sem sobrescrevê-lo.
open("sobrepython.txt", "x")
# Abre o arquivo "sobrepython.txt" no modo de criação exclusiva ("x"), que cria um novo arquivo e falha se o arquivo já existir.
open("sobrepython.txt", "b")
# Abre o arquivo "sobrepython.txt" no modo binário ("b"), que é usado para ler ou escrever arquivos em formato binário, como imagens ou arquivos executáveis.
open("sobrepython.txt", "t")
# Abre o arquivo "sobrepython.txt" no modo de texto ("t"), que é o modo padrão para arquivos de texto e permite ler ou escrever strings.
open("sobrepython.txt", "w+")
# Abre o arquivo "sobrepython.txt" no modo de leitura e escrita ("w+"), que permite ler e escrever no arquivo, mas também sobrescreve o conteúdo existente.
open("sobrepython.txt", "r+")
# Abre o arquivo "sobrepython.txt" no modo de leitura e escrita ("r+"), que permite ler e escrever no arquivo sem sobrescrever o conteúdo existente.
open("sobrepython.txt", "a+")
# Abre o arquivo "sobrepython.txt" no modo de leitura e anexação ("a+"), que permite ler e escrever no arquivo, mas adiciona o conteúdo ao final do arquivo sem sobrescrevê-lo.
open("sobrepython.txt", "x+")
# Abre o arquivo "sobrepython.txt" no modo de criação exclusiva e leitura/escrita ("x+"), que cria um novo arquivo e permite ler e escrever, mas falha se o arquivo já existir.
open("sobrepython.txt", "rt")
# Abre o arquivo "sobrepython.txt" no modo de leitura de texto ("rt"), que é equivalente a "r" e é usado para ler arquivos de texto.
```
