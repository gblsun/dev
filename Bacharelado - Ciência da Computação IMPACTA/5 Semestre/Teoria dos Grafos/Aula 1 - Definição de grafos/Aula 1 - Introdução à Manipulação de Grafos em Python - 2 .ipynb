```markdown
## Descrição do Algoritmo

Este algoritmo tem como objetivo gerar e analisar um grafo aleatório a partir de
dados simulados, utilizando as bibliotecas NetworkX, Pandas, NumPy e Matplotlib.

Inicialmente, o usuário informa a quantidade de vértices do grafo. A partir desse
valor, é criada uma lista de vértices e, em seguida, uma tabela de dados (`DataFrame`)
contendo pares de vértices de origem e destino, gerados aleatoriamente.

Esses dados são então convertidos em um grafo por meio da função
`from_pandas_edgelist`, que constrói a estrutura de conexões a partir da tabela.

Após a criação do grafo, é realizada sua visualização gráfica, permitindo observar
a distribuição das conexões entre os vértices.

Na etapa seguinte, o algoritmo calcula o **grau de cada vértice** e exibe um
**histograma da distribuição de graus**, possibilitando a análise da conectividade
da rede.

Por fim, o algoritmo identifica os **cliques presentes no grafo** e imprime tanto
a estrutura do grafo quanto uma amostra dos dados utilizados, servindo como base
para estudos sobre redes aleatórias e análise estrutural de grafos.
```


````python
# Importações
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
````

````python
# Entrada do usuário e vértices

N = int(input("Número de vértices: "))
vertices = list(range(N + 1))
````

````python
# Criação das arestas aleatórias

df = pd.DataFrame(
    {"De": np.random.choice(vertices, N), "Para": np.random.choice(vertices, N)}
)
````

````python
# Construção e desenho do grafo

G = nx.from_pandas_edgelist(df, source="De", target="Para")
nx.draw(G, node_color="Blue", with_labels=True, font_weight="bold")
plt.show()
````

````python
# Histograma dos graus

#   k: grau do grafo como um todo
#   t: grau de cada vértice

plt.hist([t for k, t in nx.degree(G)], color="red")
plt.show()
````

````python
# Cliques e visualização de dados

cliques = nx.find_cliques(G)
print(G)
print(df.head())
````
