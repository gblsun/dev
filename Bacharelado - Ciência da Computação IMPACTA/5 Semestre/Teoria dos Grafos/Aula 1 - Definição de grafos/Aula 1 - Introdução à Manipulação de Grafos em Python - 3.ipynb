```python
# Importações
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Criação do Grafo e Definição das Arestas

# Criação do Grafo
G = nx.Graph()

# Definindo as arestas (Origem, Destino, Distância)
edges = [
    ("SP", "RJ", 358), ("SP", "DF", 873), ("SP", "BELEM", 2468),
    ("SP", "POA", 1313), ("POA", "CURITIBA", 724), ("CURITIBA", "FLORIPA", 306),
    ("FLORIPA", "SP", 698), ("CUIABA", "SP", 1531), ("CURITIBA", "SP", 698),
    ("SP", "BH", 580), ("BH", "DF", 733)
]
G.add_weighted_edges_from(edges)
```

```python
# Definição das Posições Geográficas

# Posições Realistas (Coordenadas Geográficas aproximadas: Longitude, Latitude)
# Nota: Longitude é o eixo X, Latitude é o eixo Y
pos_realista = {
    "SP": (-46.63, -23.55),
    "RJ": (-43.17, -22.90),
    "DF": (-47.88, -15.79),
    "BELEM": (-48.49, -1.45),
    "POA": (-51.21, -30.03),
    "CURITIBA": (-49.27, -25.42),
    "FLORIPA": (-48.54, -27.59),
    "CUIABA": (-56.09, -15.60),
    "BH": (-43.93, -19.91)
}
```

```python
# Cálculo das Características do Grafo

# Cálculos de Características do Grafo
graus = dict(G.degree()) # Dicionário com o grau de cada vértice
```

```python
# Configuração da Figura

# Configuração da figura
plt.figure(figsize=(12, 8))
```

```python
# Desenho dos Nós

# Desenho dos Nós (Tamanho baseado no Grau)
# Quanto mais conexões a cidade tem, maior o nó
nx.draw_networkx_nodes(
    G, 
    pos_realista,
    node_size=[v * 600 for v in graus.values()],
    node_color="skyblue",
    edgecolors="navy"
)
```

```python
# Desenho das Arestas

# Desenho das Arestas
nx.draw_networkx_edges(G, pos_realista, width=2, alpha=0.6, edge_color="gray")
```

```python
# Rótulos dos Nós e das Arestas

# Rótulos das Cidades
nx.draw_networkx_labels(G, pos_realista, font_size=10, font_weight="bold")

# Rótulos das Distâncias (Pesos) nas Arestas
edge_labels = nx.get_edge_attributes(G, "weight")
nx.draw_networkx_edge_labels(G, pos_realista, edge_labels=edge_labels, font_size=8)
```

```python
# Finalização e Exibição do Grafo

plt.title("Grafo de Cidades Brasileiras (Posições Geográficas Reais)", fontsize=15)
plt.axis("off") # Remove os eixos do gráfico para limpeza visual
plt.show()
```

```python
# Exibição das Características no Console

# Exibindo características no console
print("Grau de cada vértice (número de conexões):")
for cidade, grau in graus.items():
    print(f"{cidade}: {grau}")
```
