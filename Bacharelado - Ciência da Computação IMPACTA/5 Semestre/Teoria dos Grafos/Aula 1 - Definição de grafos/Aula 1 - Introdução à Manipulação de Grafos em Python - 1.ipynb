## Descrição do Algoritmo

Este algoritmo tem como objetivo modelar, visualizar e analisar um grafo não dirigido
utilizando a linguagem Python e a biblioteca NetworkX.

O grafo é construído por meio da definição manual de suas arestas, permitindo representar
explicitamente as conexões entre os vértices. Após sua criação, é realizada a visualização
gráfica com o auxílio da biblioteca Matplotlib, possibilitando a análise da estrutura da rede.

Além da representação visual, o algoritmo calcula e exibe propriedades fundamentais do grafo,
tais como:

- Número de vértices  
- Número de arestas  
- Grau máximo dos vértices  
- Grau mínimo dos vértices  

Essas métricas permitem avaliar o nível de conectividade do grafo e compreender melhor
sua organização, servindo como base para estudos mais avançados em Teoria dos Grafos e
Análise de Redes.

```python
# Importações
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Criação do grafo

G = nx.Graph()  # Digraph

# Pode simplificar:
# elist=[(1,2),(2,6),(1,6),(6,7),(7,1),(2,7),(1,9),(9,2)]
# G.add_edges_from(elist)
G.add_edge(1, 2)
G.add_edge(2, 6)
G.add_edge(1, 6)
G.add_edge(6, 7)
G.add_edge(7, 1)
G.add_edge(2, 7)
G.add_edge(1, 9)
G.add_edge(9, 2)
```

```python
# Desenho do grafo

nx.draw(G, with_labels=True, node_color="magenta", node_size=700)
plt.show()
```

```python
# Graus

V = nx.number_of_nodes(G)
A = nx.number_of_edges(G)

print("O número de vértices presente no grafo é:", V)
print("O número de arestas presente no grafo é:", A)
```
