# Algoritmo de Floyd-Warshall

## Descrição
O algoritmo de Floyd-Warshall encontra os menores caminhos entre todos os pares de vértices em um grafo dirigido e ponderado.

## Funcionamento
A cada iteração k, o algoritmo verifica se o caminho entre i e j pode ser melhorado passando por k.

d[i][j] = min(d[i][j], d[i][k] + d[k][j])

## Estruturas
#### Matriz de distâncias

#### Matriz de Distâncias — Floyd-Warshall

A matriz de distâncias inicial é definida como:

$$
D^{(0)} = W = (w_{ij})
$$

onde:

$$
d_{ij}^{(0)} =
\begin{cases}
0, & \text{se } i = j \\
w_{ij}, & \text{se existe aresta de } i \text{ para } j \\
\infty, & \text{caso contrário}
\end{cases}
$$

A cada iteração $k$, a matriz é atualizada por:

$$
d_{ij}^{(k)} =
\min \left(
d_{ij}^{(k-1)},
d_{ik}^{(k-1)} + d_{kj}^{(k-1)}
\right)
$$
$D^{(0)}$ → matriz inicial (entrada do usuário)
$d_{ij}$ → distância de $i$ até $j$
$\infty$ → quando não existe caminho
A fórmula de baixo é o coração do algoritmo

A cada iteração,o algoritmo  
atualiza a matriz de distâncias mínimas,  
levando em conta os caminhos que passam  
por vértices intermediários já processados.  

#### Matriz de Precedência — Floyd-Warshall

Para o algoritmo Floyd-Warshall, montamos duas matrizes:

- Matriz $D$: distância entre dois vértices  
- Matriz $\pi_{ij}^{(k)}$: matriz de precedência  

A matriz de precedência inicial é definida por:

$$
\pi_{ij}^{(0)} =
\begin{cases}
\text{NIL}, & \text{se } i = j \text{ ou } w_{ij} = \infty \\
i, & \text{se } i \neq j \text{ e } w_{ij} < \infty
\end{cases}
$$

onde:

- $i$: vértice $i$ do grafo  
- $w_{ij}$: peso da aresta de $i$ para $j$

#### Limitação
Não funciona corretamente com ciclos negativos.

#### Objetivo
Calcular:
- Menores distâncias
- Caminhos mínimos

## Pseudocódigo

n = W.linhas
D(0) = W

para k = 1 até n:
    seja D(k) uma nova matriz n x n
    
    para i = 1 até n:
        para j = 1 até n:
            d_ij(k) = min(
                d_ij(k-1),
                d_ik(k-1) + d_kj(k-1)
            )

retorne D(n)

## Entrada e Exemplo — Floyd-Warshall

### Formato de entrada

O grafo é representado por uma matriz $D$, onde:

- $d_{ij}$: custo de $i$ até $j$
- $0$: mesmo vértice
- número: peso da aresta
- `inf`: não existe aresta

O uso de `inf` (infinito) indica ausência de caminho direto e permite que o algoritmo encontre caminhos indiretos mais curtos.

---

### Exemplo de entrada

Número de vértices:
4  
Matriz:  

0 3 inf 7  
8 0 2 inf  
5 inf 0 1  
2 inf inf 0  


---

### Resultado esperado

$$
\begin{bmatrix}
0 & 3 & 5 & 6 \\
5 & 0 & 2 & 3 \\
3 & 6 & 0 & 1 \\
2 & 5 & 7 & 0
\end{bmatrix}
$$

---

### Exemplo de caminho

De $0$ até $3$:  

[0, 1, 2, 3]   

Custo:  6


```python
import math
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def floyd_warshall(W):
    n = len(W)
    
    # D(0) = W
    dist = [row[:] for row in W]
    
    # Matriz de precedência
    pred = [[None if i == j or W[i][j] == math.inf else i for j in range(n)] for i in range(n)]
    
    for k in range(n):
        for i in range(n):
            for j in range(n):
                if dist[i][k] + dist[k][j] < dist[i][j]:
                    dist[i][j] = dist[i][k] + dist[k][j]
                    pred[i][j] = pred[k][j]

    return dist, pred
```

```python
def reconstruir_caminho(pred, i, j):
    if pred[i][j] is None:
        return None
    
    caminho = [j]
    
    while j != i:
        j = pred[i][j]
        caminho.insert(0, j)
    
    return caminho
```

```python
def ler_matriz():
    n = int(input("Número de vértices: "))
    
    W = []
    for i in range(n):
        linha = input(f"Linha {i}: ").split()
        
        nova_linha = []
        for valor in linha:
            if valor.lower() == "inf":
                nova_linha.append(math.inf)
            else:
                nova_linha.append(float(valor))
        
        W.append(nova_linha)
    
    return W
```

```python
def criar_grafo(W):
    G = nx.DiGraph()
    
    n = len(W)
    
    for i in range(n):
        for j in range(n):
            if W[i][j] != math.inf and i != j:
                G.add_edge(i, j, weight=W[i][j])
    
    return G
```

```python
def desenhar_grafo(G):
    pos = nx.spring_layout(G)
    
    labels = nx.get_edge_attributes(G, 'weight')
    
    nx.draw(G, pos, with_labels=True)
    nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
    
    plt.show()
```

```python
W = ler_matriz()

G = criar_grafo(W)
desenhar_grafo(G)

dist, pred = floyd_warshall(W)

print("Matriz final de distâncias:")
for linha in dist:
    print(linha)
```

```python
i = int(input("Origem: "))
j = int(input("Destino: "))

print("Caminho:", reconstruir_caminho(pred, i, j))
```

Complexidade  
- Tempo: O(n³)  
- Espaço: O(n²)  

---

- O algoritmo usa três laços (k, i, j), por isso é O(n³).   
- Usa matrizes n×n, por isso é O(n²).
