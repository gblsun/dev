## Carregamento do Dataset

O dataset está em formato CSV com delimitador ";".

```python
# Carregando o dataset

df = pd.read_csv("iris.csv", sep=";")
# Visualizar primeiras linhas
df.head()
```

## Frequência Absoluta

A **frequência absoluta** representa o número de vezes que uma determinada categoria aparece no conjunto de dados.

A fórmula da frequência absoluta é:

$$
n_i
$$

onde:

- $n_i$ representa o número de ocorrências da categoria $i$ no conjunto de dados.

```python
# Frequência absoluta

freq_absoluta = df["Species"].value_counts()

freq_absoluta
```

### Resultado Esperado:

setosa        50  
versicolor    50  
virginica     50  

```python
# Verificação das Colunas do Dataset

df.columns
```

## Frequência Relativa

A **frequência relativa** representa a proporção da frequência absoluta em relação ao número total de observações do conjunto de dados.

A fórmula da frequência relativa é dada por:

$$
f_i = \frac{n_i}{N}
$$

onde:

- $f_i$ representa a frequência relativa da categoria $i$
- $n_i$ representa a frequência absoluta da categoria $i$
- $N$ representa o número total de observações no conjunto de dados

```python
# Frequência relativa

freq_relativa = df["Species"].value_counts(normalize=True)

freq_relativa
```

### Resultado Esperado:

setosa        0.3333  
versicolor    0.3333  
virginica     0.3333  

## Tabela de Frequência Completa

Agora vamos criar uma tabela contendo:

- Frequência Absoluta
- Frequência Relativa

### Tabela de Frequência

**Frequência Absoluta**

$$
n_i
$$

Número de ocorrências da categoria \(i\).

**Frequência Relativa**

$$
f_i = \frac{n_i}{n}
$$

onde:

- \(n_i\) = frequência absoluta da categoria \(i\)
- \(n\) = número total de observações

**Frequência Relativa Percentual**

$$
f_i(\%) = f_i \times 100
$$

```python
# Construindo tabela de frequência

freq_abs = df["Species"].value_counts()
freq_rel = df["Species"].value_counts(normalize=True)

tabela_freq = pd.DataFrame({
    "Frequência Absoluta": freq_abs,
    "Frequência Relativa": freq_rel
})

tabela_freq["Frequência Relativa (%)"] = (freq_rel * 100).round(2)

tabela_freq
```

## Medidas Centrais e de Dispersão

Agora vamos calcular estatísticas descritivas para as variáveis numéricas do dataset:

- Média
- Mediana
- Desvio Padrão
- Variância


```python
# Selecionando apenas colunas numéricas
dados_numericos = df.select_dtypes(include="number")

# Média
media = dados_numericos.mean()

# Mediana
mediana = dados_numericos.median()

# Desvio padrão
desvio_padrao = dados_numericos.std()

# Variância
variancia = dados_numericos.var()

estatisticas = pd.DataFrame({
    "Média": media,
    "Mediana": mediana,
    "Desvio Padrão": desvio_padrao,
    "Variância": variancia
})

estatisticas
```

### Média do comprimento da pétala por espécie

Agora calculamos a média da variável **Petal.Length** separada por espécie.


```python
media_petala = df.groupby("Species")["Petal.Length"].mean()

media_petala
```

Desafio: Calcule a média do comprimento da pétala separada por espécie. Qual
espécie possui a maior pétala?  

Resposta: A espécie **virginica** apresenta o maior comprimento médio de pétala.


## Visualização de Dados

Os gráficos ajudam a identificar padrões e distribuições nos dados.


Os gráficos revelam padrões que os números sozinhos podem esconder. Gere as
seguintes visualizações:  
- Gráfico de Barras: Para mostrar a contagem de cada espécie.  
- Gráfico de Pizza (Pie Chart): Para visualizar a proporção das espécies no
conjunto total.  
- Histograma: Escolha a variável petal_length e observe a distribuição. Você
consegue notar algum "buraco" ou separação clara nos dados?  
- Boxplot: Gere um boxplot da largura da sépala (sepal_width) agrupado por
espécie.  

```python
# Gráfico de Barras (contagem de espécies)
import matplotlib.pyplot as plt

freq_abs.plot(kind="bar")

plt.title("Contagem de cada espécie")
plt.xlabel("Espécie")
plt.ylabel("Frequência")
plt.show()
```

```python
# Gráfico de Pizza

freq_abs.plot(kind="pie", autopct="%1.1f%%")
plt.title("Proporção das espécies")
plt.ylabel("")
plt.show()
```

```python
# Histograma — Petal Length
df["Petal.Length"].hist(bins=20)

plt.title("Distribuição do comprimento da pétala")
plt.xlabel("Petal Length")
plt.ylabel("Frequência")
plt.show()
```

Observa-se uma separação clara na distribuição do comprimento da pétala,  
indicando diferenças entre as espécies.


```python
# Boxplot — Sepal Width por espécie
df.boxplot(column="Sepal.Width", by="Species")

plt.title("Sepal Width por espécie")
plt.suptitle("")
plt.xlabel("Espécie")
plt.ylabel("Sepal Width")
plt.show()
```
