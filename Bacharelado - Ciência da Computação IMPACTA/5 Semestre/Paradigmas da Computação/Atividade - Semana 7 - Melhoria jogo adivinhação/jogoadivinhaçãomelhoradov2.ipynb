- Ranking de jogadores

## 📥 Importação de bibliotecas

```python
import random
import customtkinter as ctk
```

## ⚙️ Configuração do tema da interface

```python
ctk.set_appearance_mode("dark")  # "light" ou "dark"
ctk.set_default_color_theme("blue")
```

## 🧠 Classe principal do jogo com interface gráfica

Aqui definimos toda a lógica do jogo junto com a interface.

```python
class App(ctk.CTk):
    def __init__(self):
        super().__init__()

        self.title("Jogo de Adivinhação")
        self.geometry("400x500")

        self.numero_secreto = random.randint(1, 100)
        self.tentativas = 0
        self.pontuacao = 100
        self.ranking = []

        # Título
        self.label_titulo = ctk.CTkLabel(self, text="🎮 Adivinhe o Número", font=("Arial", 20))
        self.label_titulo.pack(pady=20)

        # Nome
        self.entry_nome = ctk.CTkEntry(self, placeholder_text="Seu nome")
        self.entry_nome.pack(pady=10)

        # Palpite
        self.entry_palpite = ctk.CTkEntry(self, placeholder_text="Digite um número (1-100)")
        self.entry_palpite.pack(pady=10)

        # Botão
        self.botao = ctk.CTkButton(self, text="Enviar Palpite", command=self.jogar)
        self.botao.pack(pady=10)

        # Feedback
        self.label_feedback = ctk.CTkLabel(self, text="")
        self.label_feedback.pack(pady=10)

        # Status
        self.label_status = ctk.CTkLabel(self, text="Tentativas: 0 | Pontos: 100")
        self.label_status.pack(pady=10)

        # Ranking
        self.text_ranking = ctk.CTkTextbox(self, height=150)
        self.text_ranking.pack(pady=10)

    def jogar(self):
        try:
            palpite = int(self.entry_palpite.get())
        except ValueError:
            self.label_feedback.configure(text="⚠️ Digite um número válido")
            return

        self.tentativas += 1
        self.pontuacao -= 10

        if palpite == self.numero_secreto:
            self.label_feedback.configure(text="🎉 Acertou!")

            nome = self.entry_nome.get() or "Anônimo"
            self.ranking.append((nome, self.pontuacao))

            self.atualizar_ranking()
            self.resetar_jogo()

        elif palpite < self.numero_secreto:
            self.label_feedback.configure(text="⬆️ Maior")
        else:
            self.label_feedback.configure(text="⬇️ Menor")

        self.label_status.configure(
            text=f"Tentativas: {self.tentativas} | Pontos: {self.pontuacao}"
        )

        self.entry_palpite.delete(0, "end")

    def atualizar_ranking(self):
        self.text_ranking.delete("1.0", "end")

        ranking_ordenado = sorted(self.ranking, key=lambda x: x[1], reverse=True)

        for i, (nome, pontos) in enumerate(ranking_ordenado, start=1):
            self.text_ranking.insert("end", f"{i}º - {nome}: {pontos}\n")

    def resetar_jogo(self):
        self.numero_secreto = random.randint(1, 100)
        self.tentativas = 0
        self.pontuacao = 100
        self.label_status.configure(text="Tentativas: 0 | Pontos: 100")
```

## 🚀 Execução do jogo

Cria a aplicação e inicia a interface gráfica.

```python
app = App()
app.mainloop()
```
